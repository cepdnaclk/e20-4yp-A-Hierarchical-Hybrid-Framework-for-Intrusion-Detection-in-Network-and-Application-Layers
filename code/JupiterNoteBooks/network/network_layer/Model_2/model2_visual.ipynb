```python
import numpy as np
import matplotlib.pyplot as plt

# Metrics
categories = ["Precision", "Recall", "F1-score", "Accuracy", "ROC-AUC"]
N = len(categories)

# Model results (Attack class metrics + accuracy + AUC)
models = {
    "LogReg":      [0.81, 0.94, 0.87, 0.89, 0.9228],
    "DecisionTree":[0.96, 0.95, 0.96, 0.97, 0.9701],
    "RandomForest":[0.98, 0.95, 0.96, 0.97, 0.9870],
    "AdaBoost":    [0.97, 0.94, 0.95, 0.96, 0.9759],
    "SVM":         [0.88, 0.94, 0.91, 0.92, 0.9627],
    "GradBoost":   [0.99, 0.94, 0.97, 0.98, 0.9809],
    "MLP":         [1.00, 0.94, 0.97, 0.98, 0.9852],
    "XGBoost":     [1.00, 0.94, 0.97, 0.98, 0.9900],
    "LightGBM":    [1.00, 0.94, 0.97, 0.98, 0.9905],
}

# Radar angles
angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(8,8))
ax = plt.subplot(111, polar=True)

for model, values in models.items():
    values = values + values[:1]
    ax.plot(angles, values, linewidth=2, label=model)
    ax.fill(angles, values, alpha=0.05)

# Axis labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

# Score range
ax.set_ylim(0.85, 1.0)

plt.title("Model 1 Performance Comparison (Radar Chart)", size=14)
plt.legend(loc="upper right", bbox_to_anchor=(1.3, 1.1))

plt.show()
```
