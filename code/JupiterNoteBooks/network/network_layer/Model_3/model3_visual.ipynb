```python
import matplotlib.pyplot as plt
import numpy as np

models = [
    "LogReg",
    "DecisionTree",
    "RandomForest",
    "XGBoost",
    "MLP",
    "LightGBM"
]

accuracy = [0.94, 0.97, 0.97, 0.97, 0.96, 0.28]
macro_f1 = [0.70, 0.91, 0.91, 0.92, 0.78, 0.19]
weighted_f1 = [0.94, 0.97, 0.97, 0.97, 0.96, 0.22]

x = np.arange(len(models))
width = 0.25

plt.figure(figsize=(12,6))

plt.bar(x - width, accuracy, width, label="Accuracy")
plt.bar(x, macro_f1, width, label="Macro F1")
plt.bar(x + width, weighted_f1, width, label="Weighted F1")

plt.xticks(x, models)
plt.ylabel("Score")
plt.title("Model 3 Performance Comparison")
plt.ylim(0,1)

plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(models, macro_f1)

plt.title("Macro F1 Score Comparison")
plt.ylabel("Macro F1")
plt.ylim(0,1)

plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

```python
import numpy as np

categories = ["Accuracy","Macro F1","Weighted F1"]
N = len(categories)

angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(8,8))
ax = plt.subplot(111, polar=True)

data = {
"logReg":[0.94, 0.70, 0.94],
"DecisionTree":[0.97,0.91,0.97],
"RandomForest":[0.97,0.91,0.97],
"XGBoost":[0.97,0.92,0.97],
"MLP":[0.96,0.78,0.96],
"LGBM":[0.28,0.19,0.22],
}

for model,values in data.items():
    values = values + values[:1]
    ax.plot(angles, values, label=model)
    ax.fill(angles, values, alpha=0.05)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

plt.title("Model Comparison Radar Chart")
plt.legend()

plt.show()
```
